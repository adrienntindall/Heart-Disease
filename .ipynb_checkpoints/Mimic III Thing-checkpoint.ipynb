# Imports/Setup

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import psycopg2
import seaborn as sns
import collections
from ggplot import *
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression

# information used to create a database connection
sqluser = 'postgres'
dbname = 'mimic'
schema_name = 'mimiciii'

# Connect to postgres with a copy of the MIMIC-III database
con = psycopg2.connect(dbname=dbname, user=sqluser, password = 'Carter2012')
```

# Querying the Data

## The following variables are taken from the dataset:
* Subject id
* Gender
* Marital Status
* Ethnicity
* Religion
* Length of stay in the ICU
* (maximum) Age at the time admitted
* Icd9 Diagnostic Codes

## The following variables were omitted for various reasons:
* Drug codes (it was too difficult to lookup what each meant, meaning drugs perscribed specifically for heart disease could slip by and unfairly bias the data. In addition, drugs administered were affecting the results much less than the diagnoces, meaning they could be easily dropped)
* Patient weight (although it would be useful and probably a valuable variable to have, not all patients have a recorded weight so around 55~70% of the data would be lost)

```python
query = """
SELECT a.subject_id, p.gender, a.marital_status, a.ethnicity, a.religion, i.los,
ROUND((cast(i.intime as date) - cast(p.dob as date))/365.242, 2) AS age
FROM admissions a
INNER JOIN patients p
ON p.subject_id = a.subject_id
INNER JOIN icustays i
ON i.subject_id = p.subject_id
;
"""
df = pd.read_sql_query(query, con)
```

```python
query = """
SELECT subject_id, icd9_code
FROM diagnoses_icd
"""
dia = pd.read_sql_query(query, con)
dia
```

# Preparing the Data

The data goes through a process to hot code variables and restrict the amount of data points to reduce run time via limiting the age range of the subjects observed.

```python
# Set cases to the data file (mostly done to make things easier to test)
cases = df
cases = cases.drop_duplicates()
cases = cases.set_index('subject_id')

## Clean Admit Data

hf = ['gender', 'marital_status', 'ethnicity', 'religion'] #Values to one hot encode
#cases = pd.get_dummies(cases, columns=hf)

gb = cases.groupby(level=0).mean() #Group rows by the index
gb.age = cases.groupby(level=0).age.max() #Take maximum age
gb['total_stays'] = cases.groupby(level=0).los.count() #Find the total amount of times someone was admitted to the ICU
gb.los = cases.groupby(level=0).los.sum() #Add up the total amount of time spent in the ICU

cases = gb

cases.loc[cases['age'] >= 90, 'age'] = 91.4 #set anyone whose age was omitted to the median age of those omitted (91.4)


## Clean Diagnostics Data

icd = dia

vc = icd['icd9_code'].value_counts() #Get frequency of each diagnoses

# Reduce the amount of features by not including diagnostic information if the total amount of diagnoces for one
#     condition/disease was less than 200

dlist = [vc.index.tolist()[x] for x in range(len(vc)) if vc[x] <= 100] 

# Set the index to the icd9_codes and drop everything contained in dlist
icd = icd.set_index('icd9_code')
icd = icd.drop(dlist, axis=0)

# dlist contains diagnoces casued by heart disease (ie. surgery, pacemakers) and are therefore not predictors of it
dlist = ['V4582', '412', 'V4582', 'V4581', 'V4501','E8790', '99672', 'V5861']
icd = icd.drop(dlist, axis=0).dropna()
icd = icd.reset_index()

# Codes for different types of heart disease
dlist = ['icd9_HD', 'icd9_9971', 'icd9_78551']
# First 2/3 digits of codes that relate to heart disease
clist = ['42', '410', '411', '412', '413', '414','391', '393', '394', '395', '396', '397', '398', '402', '403', '404', '416']

# Replacing heart disease codes with the more general 'HD'
torepl = []
for x in icd['icd9_code']:
    if(type(x) != str):
        pass
    elif (x[:2] in clist or x[:3] in clist):
        if(x not in torepl):
            torepl.append(x)
            
for x in torepl:
    icd = icd.replace(x, 'HD')

l = icd['icd9_code'].unique()
l = np.delete(l, 1)
l = [x for x in l if x is not None]
l = np.sort(l)    
    
icd = icd.set_index('subject_id')

# OHEing icd9_codes into features
icd = pd.get_dummies(icd, columns=['icd9_code'], prefix='icd9').groupby(level=0).max()

# Setting true if the patient has a form of heart disease or flase otherwise
icd['Heart_Diseased'] = icd.filter(dlist, axis=1).any(1)
icd = icd.drop(dlist, axis=1)

# Updating cases to include diagnostics information
cases = pd.concat([cases, icd], axis=1, join_axes=[cases.index]).dropna()
```

```python
ggplot(cases, aes(x='Heart_Diseased', weight='icd9_4019')) + geom_bar(fill='tomato') + ggtitle("Patients with Hypertension")
```

# Models
## Random Forest Classifier
The data is split into train and test sets and fit to multiple classifiers to see which one produces the highest true positive/true negative percentages. The first model used is the Random Forest Classifier, which uses bootstrapping to create decision trees that are based off of random sub-sections of the train set. The results from each tree are then averaged together to predict true or false (ie to predict whether the patient died or not)

```python
train, test = train_test_split(cases.dropna(), test_size=.2)
cases
```

```python
# Split data into train/test sets and fit the random forest model
rfc = RandomForestClassifier(n_estimators=500)

dlist = ['Heart_Diseased']

Xtr = train.drop(dlist, axis='columns')
ytr = train['Heart_Diseased'].astype(int) 
rfc.fit(Xtr, ytr)
```

```python
# Get important features and make a graph to display them
imp = rfc.feature_importances_
indices = np.argsort(imp)[::-1]
indices = indices[0:5]
std = np.std([rfc.feature_importances_ for tree in rfc.estimators_],
             axis=0)
plt.figure()
plt.title("Feature importances")
plt.bar(range(5), imp[indices],
       color="r", yerr=std[indices], align="center")
plt.xticks(range(5), indices)
plt.xlim([-1, 5])
ft = list(Xtr)
c = [ft[x]for x in indices] 
plt.xlabel(c)
plt.show()
```

### Brief discussion on some important features
Moving right past some of the obvious features, there are some interesting ones that became good predictors of heart disease. For instance, the number 3 feature, `icd9_V053` represents a diagnoces in which someone hasn't recived a hepatitis immunization. Why this could be at the top and accounts for more than 1% of prediction is hard to say without further research, but it could be due to a few things, such as someone not immunizing themselves because they don't take good care of their bodies or either the precense of something in the hepatitis vaccine that reduces risk of heart disease or something that hepatitis does that increases the risk of heart disease. Aside from that, most of the other icd9 codes near the top seem to be indicitive of high cholesteral/other cardiovascular problems like high blood pressure, and make sense as to why they're up there. The two `marital_status` variables are also probably there because they could be linked to age (as more older people tend to be married than younger people, and heart disease risk increases with age). Strangely, being Catholic also increases the risk of heart disease. This could be chopped up to either cultural differences between Catholics and non-Catholics involving diets/levels of exercise or simply because the data was taken entirely from the United States, a prodominetly Christain country, meaning that more Catholics die because there are more of them. Similar lines of reasoning can be attributed to whites being more at risk for heart disease, if the difference isn't genetic (as people who share an ethnicity also tend to share similar genes). Genetic or cultural differences could also explain why Hispanic/Latinos are at a much lower risk of heart disease as well. `los` could be biasing the data as it takes into considering the time spent being in the hospital for heart disease related reasons, and thus wouldn't be a good indicator if used on subject data from someone at great risk for developing heart disease but hasn't been admitted to the ICU. However, this bias is likely to be minimal (despite the high influence of `loc`) as, when excluding los, the model maintains a similar performance.
